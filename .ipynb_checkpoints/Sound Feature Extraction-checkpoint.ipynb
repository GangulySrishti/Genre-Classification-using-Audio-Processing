```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import librosa
```

```python
#loading the audio file into timeseriesarray x and samplingrate sr using defalut sr of 22100
audio_data="C:/Users/Srishti Ganguly/Documents/Workspace/4th year project/data/genres_original/blues/blues.00000.wav"
x, sr = librosa.load(audio_data, 44100)
print(x)
print(len(x))
```

```python
#extracting the zerocrossingrate feature
zcr=librosa.feature.zero_crossing_rate(x)
print(zcr)
avgzcr = np.mean(zcr)
stdzcr = np.std(zcr)
print(sum(zcr[0]), avgzcr, stdzcr)
mydict = [{'Name':'blues.00000','zcr_mean':avgzcr,'zcr_std':stdzcr}]
mydict.append({'Name':'blues.00002', 'zcr_mean':avgzcr, 'zcr_std':stdzcr})
plt.figure(figsize=(20, 5))
plt.plot(x[0:100])
plt.grid()
```

```python
#writingfileintocsv
import csv
Field = ['Name', 'zcr_mean', 'zcr_std']
with open('mycsv.csv','w') as csvfile:
    writerobj = csv.DictWriter(csvfile, fieldnames = Field)
    writerobj.writeheader()
    writerobj.writerow(mydict)
```
